```python
import numpy as np
import pandas as pd

data=pd.read_csv('BA-Final-Data-Skytrax.csv')
data
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['Unnamed: 0','Staff Service','Value for Money','Wifi & Connectivity'],inplace=True)
```

```python
#filling the null values of aircraft data with some other values might impact my data integrity
#My current focus is to get the Customer sentiment and quality service
# So dropping the rows will not effect my data instead helps me to get the meaningful insights.

data.dropna(subset=['Aircraft'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data.mode()
```

```python
data.dropna(subset=['Inflight Entertainment'], inplace=True)

data.isnull().sum()


```

```python
data
```

```python
data['Ground Service']=data['Ground Service'].fillna(data['Ground Service'].mode()[0])
data['Food & Beverages']=data['Food & Beverages'].fillna(data['Food & Beverages'].mode()[0])
data.isnull().sum()
```

```python
data.shape
```

```python
# List of columns to calculate the average
rating_columns = ['Food & Beverages', 'Inflight Entertainment', 'Seat Comfort',
                   'Cabin Staff Service', 'Ground Service', 'Value For Money']

# Calculate the row-wise average while ignoring null values
data['Overall Rating'] = data[rating_columns].mean(axis=1).round(0)
```

```python
data.shape
```

```python
data
```

```python
#to get the place from reviewer_details
# Use str.split to extract place by splitting based on '(' and ')'
data['Author'] = data['reviewer_detail'].str.split('(', expand=True)[0]
data['Date'] = data['reviewer_detail'].str.split(')', expand=True)[1]
data['Place'] = data['reviewer_detail'].str.split('(', expand=True)[1].str.split(')', expand=True)[0]
```

```python
data.head()
```

```python
data.drop(columns=['reviewer_detail'],inplace=True)
```

```python
data
```

```python
from math import nan
data['Trip Verified'] = data['reviews'].str.split('|', expand=True)[0]

data['Trip Verified'] = data['Trip Verified'].where(data['reviews'].str.contains(r'\|'), nan)

data
```

```python
data['reviews'] = data['reviews'].str.split('|').str[1].fillna(data['reviews'])
```

```python
data
```

```python
data['Trip Verified'] = data['Trip Verified'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
data.head()
```

```python
data.rename(columns={'Inflight Entertainment': 'Entertainment'}, inplace=True)
data.rename(columns={'Food & Beverages': 'Food'}, inplace=True)
```

```python
#reviews,	Food & Beverages,	Inflight Entertainment,	Seat Comfort,	Type Of Traveller,	Seat Type,	Route,	Date Flown
#Cabin Staff Service,	Ground Service,	Value For Money,	Recommended,	Aircraft,	Overall Rating,	Author,	Date,	Place,	Trip Verified
columns_order = ['Author', 'reviews','Date', 'Place', 'Aircraft', 'Type Of Traveller', 'Seat Type', 'Route', 'Date Flown',
                 'Overall Rating',  'Food', 'Entertainment', 'Seat Comfort', 'Cabin Staff Service', 'Ground Service', 'Value For Money',
                 'Trip Verified', 'Recommended']
data = data.reindex(columns=columns_order)
```
